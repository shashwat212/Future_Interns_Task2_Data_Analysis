```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
df.groupby("Category")["Sales"].sum().plot(
    kind="bar",
    title="Sales by Category",
    xlabel="Category",
    ylabel="Total Sales"
)
plt.show()
```

```python
df.groupby("Region")["Profit"].sum().plot(
    kind="pie",
    autopct="%1.1f%%",
    title="Profit by Region"
)
plt.ylabel("")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df.groupby(df["Date"].dt.month)["Sales"].sum().plot(
    kind="line",
    marker="o",
    title="Monthly Sales Trend",
    xlabel="Month",
    ylabel="Total Sales"
)
plt.show()
```

```python
print("Key Insights:")
print("1. Electronics category generates the highest sales.")
print("2. North region contributes the maximum profit.")
print("3. Sales vary across different months, showing seasonal trends.")
```
